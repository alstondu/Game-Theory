### Q1: Matrix Game (1)

Player 1 and 2 each choose a positive integer up to K (include K). If the players choose the same number, then

- player 2 pays **\$2** to player 1 if both of them choose 1
- player 2 pays **\$1** to player 1 if both of them choose an integer larger than 1

otherwise no payment is made. Each player's perferences are represented by her expected monetary payoff.

(a) When $K = 2$, write down the payoff matrix of the game, and steyp by step find any Nash equilibrum point. **[3 Marks]**

------
<font color="red">Answer</font>

(a) When $K = 2$, P2 pays **\$2** to player 1 if both of them choose 1,  P2 pays **\$1** to player 1 if both of them choose 2, otherwise, no payment is made. Therefore, the payoff matrix can be derived as:


| P1 \ P2 |    1    |    2    |
|---------|---------|---------|
| **1**   | (2, -2) | (0, 0)  |
| **2**   | (0, 0)  | (1, -1) |

A Nash equilibrium occurs when neither player has an incentive to deviate from their strategy, given the strategy of the other player. Below are the steps to find it:

- (1,1): If P1 chooses 1, P2's best response is to choose 2, as choosing 1 would result in paying $2 to P1. Therefore (1,1) is not a Nash equilibrum point;

- (1,2): If P1 chooses 1, P2's best response is to choose 2. However, If P2 chooses 2, P1's best response is also to choose 2, as choosing 2 would result in earning $1 to from P2.Therefore (1,2) is not a Nash equilibrum point;

- (2,2): If P1 chooses 2, P2's best response is to choose 1, as choosing 2 would result in paying $1 to P1. Therefore (2,2) is not a Nash equilibrum point;

- (2,1): If P1 chooses 2, P2's best response is to choose 1. However, If P2 chooses 1, P1's best response is also to choose 1, as choosing 1 would result in earning $2 to from P2.Therefore (2,1) is not a Nash equilibrum point;

Therefore, this game has no there is no pure strategy Nash equilibrum.

In terms of mixed strategy, assume $\alpha$ is the probability of P1 selecting 1, and $1-\alpha$ is the probability of choosing 2. Similarly, assume $\beta$ to be the strategy for P2.
The Nash equilibrum is reached when P2 selects a strategy so that P1's payoff for selecting 1 equals that for selecting 2:
$$2\times\beta = 1\times (1-\beta)$$
Similarly, the strategy chosen by P1 makes P2's payoff for selecting 1 equals that for selecting 2:
$$-2\times\alpha = (-1)\times(1 - \alpha)$$
Solving the two equations above, get $\alpha = 1/3$, $\beta = 1/3$, and the mixed-strategy Nash equilibrum is (1/3, 2/3), (1/3, 2/3).

```
import nashpy
import numpy as np
import itertools

def find_nash_equilibria_2x3(A, B):
    # Initialize an empty list to store the Nash equilibria
    results = []
    # Define the dimensions of the game
    M, N = 2, 3
    # Iterate over all possible combinations of strategies for both players
    for k in range(1, 3):
        for I in itertools.combinations(range(M), k):
            for J in itertools.combinations(range(N), k):
                # Convert the combinations to lists
                I, J = list(I), list(J)
                # Extract the subsets of the payoff matrices
                # corresponding to the current combinations of strategies
                A_subset = A[np.ix_(I, J)]
                B_subset = B[np.ix_(I, J)]

                # Define the right-hand side of the system of linear equations
                b = np.concatenate([np.zeros(k), np.ones(1)])
                # Define the system of linear equations for player 1
                constraint = np.block([[B_subset.T, -np.ones((k, 1))],
                                    [np.ones((1, k)), np.zeros((1, 1))]])
                # Solve the system to find player 1's mixed strategy
                x_support = np.linalg.solve(constraint, b)[:-1]

                # Define the system of linear equations for player 2
                constraint = np.block([[A_subset, -np.ones((k, 1))],
                                    [np.ones((1, k)), np.zeros((1, 1))]])
                # Solve the system to find player 2's mixed strategy
                y_support = np.linalg.solve(constraint, b)[:-1]

                # Initialize the mixed strategies for both players
                x = np.zeros(M)
                y = np.zeros(N)

                # Update the mixed strategies with the solutions found
                x[I] = x_support
                y[J] = y_support

                # Check if the mixed strategies are valid
                if all(x >= 0) and all(y >= 0):
                    # Compute the expected payoffs for both players
                    x_expected_payoff = A.dot(y)
                    y_expected_payoff = B.T.dot(x)

                    # Check if the supports are best responses to each other
                    if np.isclose(x_expected_payoff[I[0]], max(x_expected_payoff)) and \
                        np.isclose(y_expected_payoff[J[0]], max(y_expected_payoff)):
                        # If they do, add them to the list of results
                        results.append((x, y))

    # Return the list of Nash equilibria
    return results

A = np.array([[4, 0, 0], [0, 2, 1]])
B = np.array([[2, 0, 1], [0, 4, 3]])
print(find_nash_equilibria_2x3(A, B))
```

```
import numpy as np
# Iterated environment
class iterated_games():
  def __init__(self, payoff1, payoff2, max_step):
    self.payoff1 = payoff1
    self.payoff2 = payoff2
    self.max_step_num = max_step
    self.reset()
  def reset(self):
    self.history = {'agent1':[], 'agent2':[]}
    self.step_num = 0
    return self.history
  def step(self, a1, a2):
    r1, r2 = self.payoff1[a1, a2], self.payoff2[a1, a2]
    self.history['agent1'].append(a1)
    self.history['agent2'].append(a2)
    self.step_num += 1
    done = False
    if self.step_num == self.max_step_num:
      done = True
    return r1, r2, self.history, done

# Implement different strategies
# Grim trigger
class grim_trigger_agent:
  def __init__(self, idx=0):
    self.idx = idx
  def step(self, history_both):
    if self.idx == 0:
      history = history_both['agent2']
    else:
      history = history_both['agent1']
    if 1 in history:
      action = 1
    else:
      action = 0
    return action

# Limited punishment
class limited_punish_agent:
  def __init__(self, k=2, idx=0): # Set k to 2
    self.k = k
    self.idx = idx
    self.punish_step = 0
    self.punish = False
    self.punish2normal = False
  def step(self, history_both):
    if self.idx == 0:
      history = history_both['agent2']
    else:
      history = history_both['agent1']
    if len(history) == 0:
      action = 0
    elif self.punish2normal == True:
      action = 0
      self.punish2normal = False
    elif history[-1] == 1 and self.punish == False:
      action = 1
      self.punish = True
      self.punish_step = self.k - 1
      if self.punish_step == 0:
        self.punish2normal = True
        self.punish = False
    elif self.punish == True:
      action = 1
      self.punish_step -= 1
      if self.punish_step == 0:
        self.punish2normal = True
        self.punish = False
    else:
      action = 0
    return action

# tit-for-tat
class tit_for_tat_agent:
  def __init__(self, idx=0):
    self.idx = idx
  def step(self, history_both):
    if self.idx == 0:
      history = history_both['agent2']
    else:
      history = history_both['agent1']
    if len(history) == 0:
      action = 0
    else:
      action = history[-1]
    return action

# All-defection
class all_defection_agent:
  def __init__(self, idx):
    self.idx = idx
  def step(self, history_both):
    action = 1
    return action

# Functions for evaluation
# Function to calculation of discounted accumulated returns
def get_discount_return(return_list, delta):
  return_all = np.array(return_list)
  discount = np.array([delta ** i for i in range(len(return_list))])
  return_all = np.sum(return_all * discount)
  return return_all

# rollout two strategies and return returns for these two strategies
def evaluate(env, agent1, agent2, delta):
  history = env.reset()
  done = False
  r1_list = []
  r2_list = []
  while True:
    if done:
      break
    a1 = agent1.step(history)
    a2 = agent2.step(history)
    r1, r2, history, done = env.step(a1, a2)
    r1_list.append(r1)
    r2_list.append(r2)
  return1 = get_discount_return(r1_list, delta)
  return2 = get_discount_return(r2_list, delta)
  return return1, return2

# Payoff matrices for the two players
payoff1 = np.array([[3,0],[5,2]])
payoff2 = np.array([[3,5],[0,2]])
delta = 0.95
max_step = 100
env = iterated_games(payoff1, payoff2, max_step)
agent_list = [grim_trigger_agent, limited_punish_agent, tit_for_tat_agent, all_defection_agent]
payoff_matrix = np.zeros([4,4])
for i in range(4):
  for j in range(4):
    env.reset()
    agent1 = agent_list[i](idx=0)
    agent2 = agent_list[j](idx=1)
    payoff_matrix[i][j] = evaluate(env, agent1, agent2, delta)[0]
print(payoff_matrix)
```

As is shown in the fugures above, the DoubleQLearningAgent can beat the mixed agent after around 20-episode training, illustrated by its cumulative reward higher than that of the mixed agent. Also, the average Q value converges as well.

## Q5 Bertrand model of oligopoly
The Bertrand model of oligopoly models a market situation in which firms (agents) compete solely on price. It assumes that each firm is a price-taker and that the market price is determined by the lowest price offered by any of the firms. The Bertrand model of oligopoly is a game of sequential moves, in which firms (agents) choose their prices first, and then the consumers choose the firm they want to buy from.

For example, consider a market with two firms (agents) selling a **homogeneous good**. Both firms have the same cost structure, and both can produce any quantity at the same cost. In this case, the market price is determined by the lowest price offered by either firm. **If one firm offers a lower price, it will capture all the market *demand*, and the other firm will not sell any units.** The game is played as follows:

1. Both firms simultaneously choose a price.
2. Given the prices, each firm's profit is calculated based on the market share it captures.
3. The firms revise their prices based on the profits they received
4. The process 1-3 repeat until it convergences to a Nash equilibrium.


(a)

We consider the following parameter setting for a two firms case, then we have the following notations:

* Players: Two firms $i=1,2$ with constant marginal cost $c\geq0$
* Strategic Variables: Firms select the price level $p_1, p_2 \geq 0$;
* Market Demand $d(p_i)$, which is determined by the price.

We can derive the demand function $D_{1}$ for Firm 1 given the price $p_1, p_2$ as:

$$
D_{1}\left(p_1, p_2\right)= \begin{cases}d\left(p_1\right), & \text { if } p_1<p_2 \\ \frac{d\left(p_1\right)}{2}, & \text { if } p_1=p_2 \\ 0, & \text { otherwise }\end{cases}
$$

Also the demand function $D_{2}$ for Firm 2 given the price $p_1, p_2$ as:

$$
D_{2}\left(p_1, p_2\right)= \begin{cases}d\left(p_2\right), & \text { if } p_2<p_1 \\ \frac{d\left(p_2\right)}{2}, & \text { if } p_2=p_1 \\ 0, & \text { otherwise }\end{cases}
$$

The reward function for both company can be calculated by:
$$
R_{i}= D_{i}(p_i - c)
$$

Assume that
$$

```
import numpy as np
import matplotlib.pyplot as plt
def demand(p):
    return 5 - p if p < 5 else 0

def reward(p1, p2, c):
    if p1 < p2:
        return demand(p1) * (p1 - c)
    elif p1 == p2:
        return demand(p1) / 2 * (p1 - c)
    else:
        return 0

def best_response(p):
    if p > 3:
        return 3
    elif 1.1 < p <= 3:
        return round(p - 0.1, 1)
    else:
        return p

# Initialize prices
p1 = np.random.choice(np.arange(0.0, 5.0, 0.1))
p2 = np.random.choice(np.arange(0.0, 5.0, 0.1))
print(f"Initial prices: p1 = {p1}, p2 = {p2}")

c1 = c2 = 1.0
max_iter = 50

# Store prices for plotting
p1_values = [p1]
p2_values = [p2]

for i in range(max_iter):
    if i % 2 == 0:
        p1_new = best_response(p2)
        p1 = p1_new
    else:
        p2_new = best_response(p1)
        p2 = p2_new
    print(f"Iteration {i+1}: p1 = {p1}, p2 = {p2}")

    # Store prices for plotting
    p1_values.append(p1)
    p2_values.append(p2)
print(f"Converged prices: p1 = {p1}, p2 = {p2}")
print(f"Final rewards: R1 = {reward(p1, p2, c1)}, R2 = {reward(p2, p1, c2)}")

# Plot price convergence
plt.figure(figsize=(10, 6))
plt.plot(p1_values, label='Price 1')
plt.plot(p2_values, label='Price 2')
plt.xlabel('Iteration')
plt.ylabel('Price')
plt.title('Price Convergence in Bertrand Oligopoly')
plt.legend()
plt.grid(True)
plt.show()
```
